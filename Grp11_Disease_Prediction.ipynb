```python
# Separating features (X) and target (y)
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

# **Create an Evaluation Function**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# Evaluation function for classification models
def evaluate_model(model_name, y_true, y_pred, class_names):
    print(f"\n==================== {model_name} ====================")

    # Accuracy
    acc = accuracy_score(y_true, y_pred)
    print("Accuracy:", acc)

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred, target_names=class_names))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    # Plot CM (normalized)
    plt.figure(figsize=(8,6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
evaluate_model("Logistic Regression", y_test, log_pred, class_names)
evaluate_model("KNN", y_test, knn_pred, class_names)
evaluate_model("SVM", y_test, svm_pred, class_names)
evaluate_model("Decision Tree", y_test, dt_pred, class_names)
evaluate_model("Random Forest", y_test, rf_pred, class_names)
evaluate_model("Naive Bayes", y_test, nb_pred, class_names)
evaluate_model("Gradient Boosting", y_test, gb_pred, class_names)

# Only if XGBoost installed
# evaluate_model("XGBoost", y_test, xgb_pred, class_names)
```

```python
# ===== Next step: hyperparameter tuning + pipeline building + save =====
# Paste and run this code in your notebook (after imports & after X_train, X_test, y_train, y_test exist)

import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import joblib
import seaborn as sns
import matplotlib.pyplot as plt

# --- 0. Ensure label encoder variable name matches your notebook ---
# You said you used 'le_prognosis' earlier; verify it exists
try:
    class_names = le_prognosis.classes_
except NameError:
    # If you don't have the encoder, rebuild it from original CSV
    orig = pd.read_csv("Modified_Dataset_Humidity_WindSpeed_Missing.csv")
    from sklearn.preprocessing import LabelEncoder
    le_prognosis = LabelEncoder()
    le_prognosis.fit(orig['prognosis'])
    class_names = le_prognosis.classes_
print("Number of classes:", len(class_names))

# --- 1. Candidate models and param grids (RandomizedSearchCV for speed) ---
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
gb = GradientBoostingClassifier(random_state=42)

rf_param_dist = {
    'n_estimators': [100, 200, 400, 800],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

gb_param_dist = {
    'n_estimators': [100, 200, 400],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [3, 5, 8],
    'subsample': [0.6, 0.8, 1.0]
}

# --- 2. Randomized search with 5-fold CV (scoring = weighted F1 for multiclass) ---
rnd_searches = {}
for name, model, param_dist in [
    ("RandomForest", rf, rf_param_dist),
    ("GradientBoosting", gb, gb_param_dist)
]:
    print(f"\nRunning RandomizedSearchCV for {name} ...")
    rsearch = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_dist,
        n_iter=20,
        scoring='f1_weighted',
        cv=5,
        random_state=42,
        n_jobs=-1,
        verbose=1
    )
    rsearch.fit(X_train, y_train)
    print(f"Best {name} CV score (weighted F1): {rsearch.best_score_:.4f}")
    print(f"Best {name} params:", rsearch.best_params_)
    rnd_searches[name] = rsearch

# --- 3. Compare best CV scores and pick best model ---
best_name = max(rnd_searches.keys(), key=lambda k: rnd_searches[k].best_score_)
best_search = rnd_searches[best_name]
print(f"\nSelected best model by CV weighted-F1: {best_name} (score {best_search.best_score_:.4f})")

# --- 4. Build preprocessing pipeline (same operations you used: median fill + scaler for continuous) ---
continuous_features = ['Age', 'Temperature (C)', 'Humidity', 'Wind Speed (km/h)']
all_features = X.columns.tolist()
binary_features = [c for c in all_features if c not in continuous_features]

continuous_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),   # handles Humidity & Wind Speed NaNs safely
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('cont', continuous_pipeline, continuous_features),
        ('bin', 'passthrough', binary_features)
    ],
    remainder='drop'
)

# --- 5. Final pipeline: preprocessor + selected best estimator (use best estimator object from RandomizedSearchCV) ---
final_model = best_search.best_estimator_
final_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', final_model)
])

# Fit the pipeline on the training set
final_pipeline.fit(X_train, y_train)

# Evaluate on test set
y_pred_test = final_pipeline.predict(X_test)
print("\nTest Accuracy:", accuracy_score(y_test, y_pred_test))
print("Test weighted F1:", f1_score(y_test, y_pred_test, average='weighted'))
print("\nClassification Report:\n", classification_report(y_test, y_pred_test, target_names=class_names))

# Confusion matrix plot (normalized)
cm = confusion_matrix(y_test, y_pred_test)
cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(10,8))
sns.heatmap(cmn, annot=True, fmt=".2f", xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.ylabel('True label'); plt.xlabel('Predicted label'); plt.title(f'{best_name} - Normalized Confusion Matrix')
plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0)
plt.show()

# --- 6. Save the final pipeline and the label encoder ---
joblib.dump(final_pipeline, "disease_prediction_pipeline.pkl")
joblib.dump(le_prognosis, "le_prognosis.pkl")
print("\nSaved: disease_prediction_pipeline.pkl and le_prognosis.pkl")
```
